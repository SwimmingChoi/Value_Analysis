```python
import numpy as np
import matplotlib.pyplot as plt

labels = ['Ach', 'Ben', 'Con', 'Hed', 'Pow', 'Sec', 
          'SD', 'Sti', 'Tra', 'Uni']
values = [1.5, 1.0, 1.5, 1.0, 4.0, 1.0, 1.0, 1.5, 1.5, 1.0]


num_vars = len(labels)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()


values += values[:1]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, values, color='yellowgreen', alpha=0.25)
ax.plot(angles, values, color='yellowgreen', linewidth=3)

ax.grid(color='lightgray', linestyle='--', linewidth=0.5)
ax.set_yticklabels(np.arange(0, 6, 1))

ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, fontsize=40, color='green', fontweight='bold')

# 차트 표시
plt.show()

```
